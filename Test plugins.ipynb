```python
import pandas as pd
from random import random

df = pd.DataFrame([
    random() for i in range(1000)
])

def import_plugin(plugin_name):
    module = __import__(plugin_name)
    if getattr(module, 'filter', None):
        print('Activate plugin %s' % plugin_name)
        return module.filter
    else:
        message = 'Plugin %s does not have filter' % plugin_name
        print(dir(module))
        print(module.__cached__)
        raise Exception(message)
    
def get_plugins():
    with open('plugins.cfg') as f:
        modules = [line.strip() for line in f.readlines()]
        return map(import_plugin, modules)

for plugin in get_plugins():
    df = plugin(df)

df
```
